```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
#import the dataset you obtained from Kaggle
df_path = ('/content/breast-cancer.csv')
```

```python
#read the csv file and display the dataframe we are working with
df = pd.read_csv(df_path)
df.head()
```

```python
#display the datatypes of each column in a Pandas DataFrame
df.dtypes
```

```python
#check to see if any duplicate rows exsist in our DataFrame
df.duplicated().sum()
```

```python
#drop unnessecary columns
df = df.drop(['id'],axis=1)
```

```python
le = LabelEncoder()
df['diagnosis'] = le.fit_transform(df['diagnosis'])
```

```python
#Overview of the Data Characteristics
print(df.info())
```

```python
#Visualizing Data Distributions
import seaborn as sns
import matplotlib.pyplot as plt

#Create a vertical box plot
for feature in df.columns:
  sns.boxplot(x = 'diagnosis', y = feature, data = df, orient='v', palette = 'Reds')
  plt.title(f'Box Plot of {feature} by Diagnosis')
  plt.show()
```

Model Experimentation (Logistic Regression)

```python
# X is created to represent the features removing the target variable in our DataFrame
X = df.drop('diagnosis', axis=1)
# y representsthe target variable
y = df['diagnosis']
```

```python
scaler = StandardScaler()
scaler.fit_transform(df)
```

```python
#split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=99)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter = 3000)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

Model Evaluation

```python
#Confusion Matrix
model_accuracy = accuracy_score(y_test, y_pred)
model_conf_matrix = confusion_matrix(y_test, y_pred)
model_classification_rep = classification_report(y_test, y_pred)
```

```python
print(f'model_accuracy: {model_accuracy: .2f}')
print(f'\nmodel_conf_matrix')
print(model_conf_matrix)
```

```python
print(f'\nmodel_classification_rep')
print(model_classification_rep)
```
